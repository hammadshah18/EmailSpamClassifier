```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
email=email[["message","detection"]]
```

```python
email
```

```python
email["category"]=email["Label"].apply(lambda x:1 if x=="spam" else 0)
```

```python
email.sample(10)
```

```python
email=email.rename(columns={"Label":"Label","message":"message","Category":"Spam"})
email
```

```python
email.columns=(["Label","message","Spam"])
```

```python
email
```

```python
x_train,x_test,y_train,y_test=train_test_split(email.message,email.category,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
v=CountVectorizer()
x_train_count=v.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train_count,y_train)
```

```python
txt_msg=["Congratulations! You have won a free lottery ticket. Claim now!",
         "Hey, are we still meeting for dinner tonight?",
         "URGENT! Call this number to claim your prize money",
         "Don’t forget to bring your notes for tomorrow’s class.",
         "Win cash prizes easily, click this link now!"]
```

```python
dummy=v.transform(txt_msg)
display(dummy.shape)
```

```python
model.predict(dummy)
```

```python
model.predict_proba(dummy)
```

```python
x_test_count=v.transform(x_test)
x_test_count.shape
```

```python
model.score(x_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([
    ("Vectorizer",CountVectorizer()),
    ("NB",MultinomialNB())
],verbose=True)




```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```
